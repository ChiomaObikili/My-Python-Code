# the results from The aggregated mean of the entire DataFrame differs from that of a subset because the subset typically contains fewer and potentially different data points, leading to a different average that reflects only that portion of the data rather than the whole.

```python
df_ords.groupby('loyalty_flag')['prices'].describe()         #performing a multiple aggregation for the entire dataframe
```

```python
df_ords.groupby('loyalty_flag')['prices'].mean()
```

```python
user_avg_price = df_ords.groupby('user_id')['prices'].mean()       # Calculate average price per user
```

```python
user_avg_price
```

```python
df_ords = df_ords.merge(user_avg_price.rename('avg_price'), on='user_id')    # Merge the average price back to the original DataFrame
```

```python
df_ords.loc[df_ords['avg_price'] < 10, 'spending_flag'] = 'Low spender'       # creating a flag
```

```python
df_ords.loc[df_ords['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords['spending_flag'].value_counts(dropna = False)    #To use on the new dataframe
```

```python
user_order_freq = df_ords.groupby('user_id')['days_since_prior_order'].median()  # Calculate median days between orders per user
```

```python
df_ords = df_ords.merge(user_order_freq.rename('median_days_since_prior_order'), on='user_id')   # Merge this median value back to the main DataFram
```

```python
df_ords.loc[df_ords['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'     # creating a flag
```

```python
df_ords.loc[(df_ords['median_days_since_prior_order'] > 10) & 
                     (df_ords['median_days_since_prior_order'] <= 20), 
                     'order_frequency_flag'] = 'Regular customer'
```

```python
df_ords.loc[df_ords['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_ords['order_frequency_flag'].value_counts(dropna = False)    #To use on the new dataframe
```
